```python
# Creating a dictionary
sample_dict = {
    'name': 'Alice',
    'age': 30,
    'location': 'New York',
    'skills': ['Python', 'Data Analysis']
}
print("Original Dictionary:", sample_dict)

# Accessing a value by key
print("Accessing 'age':", sample_dict.get('age', "Key not found"))

# Adding a new key-value pair
sample_dict['job'] = 'Engineer'
print("After Adding 'job':", sample_dict)

# Updating an existing key's value
sample_dict['location'] = 'San Francisco'
print("After Updating 'location':", sample_dict)

# Deleting a key-value pair by key
sample_dict.pop('age', "Key not found")
print("After Deleting 'age':", sample_dict)

# Retrieving all keys
keys = list(sample_dict.keys())
print("All Keys:", keys)

# Retrieving all values
values = list(sample_dict.values())
print("All Values:", values)

# Retrieving all key-value pairs as a list of tuples
items = list(sample_dict.items())
print("All Items:", items)

# Checking if a key exists in the dictionary
key_exists = 'name' in sample_dict
print("Does 'name' exist in the dictionary?", key_exists)

# Copying a dictionary
sample_dict_copy = sample_dict.copy()
print("Copy of the Dictionary:", sample_dict_copy)

# Removing and returning a specific value (if the key exists)
job = sample_dict.pop('job', "Key not found")
print("After Popping 'job':", job)
print("Dictionary after removing 'job':", sample_dict)

# Removing and returning the last key-value pair added
if sample_dict:
    last_entry = sample_dict.popitem()
    print("Popped Last Entry:", last_entry)
else:
    print("Dictionary is empty")
print("Dictionary after popping the last entry:", sample_dict)

# Setting a default value for a key if it does not exist
country = sample_dict.setdefault('country', 'USA')
print("After Setting default for 'country':", sample_dict)

# Merging another dictionary into the existing one
sample_dict.update({'hobby': 'Reading', 'age': 28})
print("After Merging with another dictionary:", sample_dict)

# Clearing all elements in the dictionary
sample_dict.clear()
print("Dictionary after Clearing:", sample_dict)
```

```python
# Creating a list
sample_list = [1, 2, 3, 4, 5, 'apple', 'banana', 'cherry']
print("Original List:", sample_list)

# Accessing an element by index
print("Accessing element at index 2:", sample_list[2])

# Adding an element to the end of the list
sample_list.append('orange')
print("After Appending 'orange':", sample_list)

# Inserting an element at a specific position
sample_list.insert(3, 'grape')
print("After Inserting 'grape' at index 3:", sample_list)

# Removing an element by value
sample_list.remove('apple')
print("After Removing 'apple':", sample_list)

# Removing an element by index
last_element = sample_list.pop()  # Removes the last element
print("After Popping Last Element:", last_element)
print("List after Popping:", sample_list)

# Removing an element at a specific index
third_element = sample_list.pop(2)  # Removes the element at index 2
print("After Popping Element at index 2:", third_element)
print("List after Removing an Element by Index:", sample_list)

# Finding the index of an element
index_of_banana = sample_list.index('banana')
print("Index of 'banana':", index_of_banana)

# Counting occurrences of an element
count_of_3 = sample_list.count(3)
print("Count of 3 in the List:", count_of_3)

# Sorting the list (only works if all elements are comparable, so let's try with a new list of numbers)
numbers_list = [5, 3, 1, 4, 2]
numbers_list.sort()
print("Sorted Numbers List:", numbers_list)

# Reversing the list
sample_list.reverse()
print("Reversed List:", sample_list)

# Copying the list
sample_list_copy = sample_list.copy()
print("Copy of the List:", sample_list_copy)

# Extending the list with another list
sample_list.extend([10, 11, 12])
print("After Extending the List:", sample_list)

# Clearing all elements from the list
sample_list.clear()
print("List after Clearing:", sample_list)
```

```python
# Creating a tuple
sample_tuple = (1, 2, 3, 'apple', 'banana', 'cherry')
print("Original Tuple:", sample_tuple)

# Accessing elements by index
print("Accessing element at index 2:", sample_tuple[2])
print("Accessing last element:", sample_tuple[-1])

# Slicing a tuple (getting a range of elements)
print("Slicing from index 1 to 4:", sample_tuple[1:4])

# Counting occurrences of an element
count_of_apple = sample_tuple.count('apple')
print("Count of 'apple' in the Tuple:", count_of_apple)

# Finding the index of an element
index_of_banana = sample_tuple.index('banana')
print("Index of 'banana':", index_of_banana)

# Concatenating two tuples
new_tuple = sample_tuple + (10, 11, 12)
print("After Concatenating with another tuple:", new_tuple)

# Repeating a tuple
repeated_tuple = sample_tuple * 2
print("Repeating the Tuple twice:", repeated_tuple)

# Checking if an element exists in the tuple
is_banana_in_tuple = 'banana' in sample_tuple
print("Is 'banana' in the Tuple?", is_banana_in_tuple)

# Finding the length of the tuple
tuple_length = len(sample_tuple)
print("Length of the Tuple:", tuple_length)

# Converting a tuple to a list (useful for modifications)
tuple_as_list = list(sample_tuple)
print("Tuple converted to List:", tuple_as_list)

# Converting a list back to a tuple
modified_tuple = tuple(tuple_as_list + ['orange'])
print("List converted back to Tuple with an additional element:", modified_tuple)
```
